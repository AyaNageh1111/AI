```python
class Environment:
    '''
    Abstract base class for an (interactive) environment formulation.
    It declares the expected methods to be used to solve it.
    All the methods declared are just placeholders that throw errors if not overriden by child "concrete" classes!
    '''
    
    def __init__(self):
        '''Constructor that initializes the problem. Typically used to setup the initial state.'''
        self.state = None
    
    def actions(self):
        '''Returns an iterable with the applicable actions to the current environment state.'''
        raise NotImplementedError
    
    def apply(self, action):
        '''Applies the action to the current state of the environment and returns the new state from applying the given action to the current environment state; not necessarily deterministic.'''
        raise NotImplementedError
    
    @classmethod
    def new_random_instance(cls):
        '''Factory method to a problem instance with a random initial state.'''
        raise NotImplementedError
```

```python
def q_learning(env, q={}, n={}, f=lambda q, n: (q+1)/(n+1), alpha=lambda n: 0.5, error=1e-6, verbose=False):
    '''Q-learning implementation that trains on an environment till no more actions can be taken'''
    if verbose: visualizer = Visualizer(env)
    while env.state is not None:
        if verbose: visualizer.visualize([env.state])
        state = env.state[0]
  #      print('ssssssssssss',state)
        action = max(env.actions(),
                     key=lambda next_action: f(q.get((state, next_action), 0), n.get((state, next_action), 0)))
        n[(state, action)] = n.get((state, action), 0) + 1
        reward = env.apply(action)
        q[(state, action)] = q.get((state, action), 0) \
                           + alpha(n[state, action]) \
                           * (reward + env.discount * max((q.get((env.state, next_action), 0) for next_action in env.actions()), default=0)
                              - q.get((state, action), 0))
       # print(env.discount)
    return q, n
```

```python
from math import inf
from time import time
from itertools import count

def simulate(env_ctor, n_iterations=inf, duration=inf, **q_learning_params):
    '''A helper function to train for a fixed number of iterations or fixed time'''
    for param in ('q', 'n'): q_learning_params[param] = q_learning_params.get(param, {})
    start_time = time()
    i = count()
    while time() < start_time + duration and next(i) < n_iterations:
        env = env_ctor()
        q, n = q_learning(env, **q_learning_params)
    return q_learning_params['q'], q_learning_params['n']
```

```python
from random import choice, randrange

ZCMap = [ 
        ["%", "%", "%","%","%","%","%","%","%","%","%","%","%","%","%","%","%","%","%","%","%","%","%","%","%","%","%","%","%","%","%","%","%","%","%","%","%","%","%","%","%","%"],
        ["-", "-", "-", "-", "-", "-", "-", "-", "-", "-", "-", "-", "-", "-", "Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt", "-", "Desesrt","Desesrt","Desesrt","Desesrt","Desesrt", "-", "-", "-", "-", "-", "-", "-", "-", "-", "-", "-", "-", "-", "-", "-", "-", ],
        ["%", "-", "Dorms", "Dorms", "Dorms", "Dorms", "Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","-" ,"-","-","-","-","-","-","-","-","-","-","-","-","-","-","-","-","-","-","-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","%"],
        ["%", "-", "Dorms", "Dorms", "Dorms", "Dorms", "Dorms", "Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Service","Service","Service","%"],
        ["%", "-", "Dorms", "Dorms", "Dorms", "Dorms", "Dorms", "Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Service","Service","Service","%"],
        ["%", "-", "Dorms", "Dorms", "Dorms", "Dorms", "Dorms", "Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Service","Service","Service","%"],
        ["%","%", "-","Dorms", "Dorms", "Dorms", "Dorms", "Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","-","-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","-","-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","-","%"],
        ["%","%","-", "Dorms", "Dorms", "Dorms", "Dorms", "Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","-","Desesrt","Desesrt","Desesrt","Desesrt","-","%","%"],
        ["%","%","-", "Desesrt", "Desesrt", "Desesrt", "Desesrt", "Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","-","Desesrt","Desesrt","Desesrt","Desesrt","-","%","%"],
        ["%","%", "%","-", "Desesrt", "Desesrt", "Desesrt", "Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","-","Desesrt","Desesrt","Desesrt","-","%","%","%"],
        ["%","%", "%", "-", "Desesrt", "Desesrt", "Desesrt", "Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","-","Desesrt","Desesrt","Desesrt","-","%","%","%"],
        ["%","%", "%","-" ,"Desesrt", "Desesrt", "Desesrt", "Desesrt","Desesrt","Desesrt","Desesrt","-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","-","Desesrt","Desesrt","Desesrt","Desesrt","-","Desesrt","Desesrt","Desesrt","-","%","%","%"],
        ["%","%", "%", "%","-" ,"Desesrt", "Desesrt", "Desesrt","Desesrt","Desesrt","-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Academic","Academic","Academic","Academic","Desesrt","Desesrt","Desesrt","Gardens","Gardens","-","Gardens","Gardens","Desesrt","-","Desesrt","Desesrt","Desesrt","-","%","%","%"],
        ["%","%", "%", "%", "%","-", "Desesrt", "Desesrt","Desesrt","Desesrt","-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Academic","Academic","Academic","Academic","Academic","Academic","Desesrt","Desesrt","Gardens","Gardens","-","Gardens","Gardens","Desesrt","-","Desesrt","Desesrt","Desesrt","-","%","%","%"],
        ["%","%", "%", "%", "%","-", "Desesrt", "Desesrt","Desesrt","-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Academic","Academic","Academic","Academic","Academic","Academic","Academic","Academic","Gardens","Gardens","-","-","Gardens","Gardens","Desesrt","-","Desesrt","Desesrt","Desesrt","-","%","%","%"],
        ["%","%", "%", "%", "%","-", "Desesrt", "Desesrt","-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Academic","Academic","Academic","Academic","Academic","Academic","Academic","Academic","Academic","Academic","Gardens","-","-","Gardens","Gardens","Desesrt","-","Desesrt","Desesrt","Desesrt","-","%","%","%"],
        ["%","%", "%", "%", "%","-", "Desesrt", "-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Academic","Academic","Academic","Academic","Academic","Academic","Academic","Academic","Academic","Academic","Academic","Academic","-","-","Gardens","Gardens","Gardens","-","Gardens","Gardens","Desesrt","-","%","%","%"],
        ["%","%", "%", "%", "%","-", "-", "Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Academic","Academic","Academic","Academic","Academic","Academic","Academic","Academic","Academic","Academic","Academic","Academic","Academic","Academic","-","Gardens","Gardens","Gardens","-","Gardens","Gardens","Desesrt","-","%","%","%"],
        ["%","%", "%", "%", "%","-", "Desesrt", "-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Academic","Academic","Academic","Academic","Academic","Academic","Academic","Academic","Academic","Academic","Academic","Academic","-","-","Gardens","Gardens","Gardens","-","Gardens","Gardens","Desesrt","-","%","%","%"],
        ["%","%", "%", "%", "%","-", "Desesrt", "Desesrt","-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Academic","Academic","Academic","Academic","Academic","Academic","Academic","Academic","Academic","Academic","Gardens","-","-","Gardens","Gardens","Desesrt","-","Desesrt","Desesrt","Desesrt","-","%","%","%"],
        ["%","%", "%", "%", "%","-", "Desesrt", "Desesrt","Desesrt","-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Academic","Academic","Academic","Academic","Academic","Academic","Academic","Academic","Gardens","Gardens","-","-","Gardens","Gardens","Desesrt","-","Desesrt","Desesrt","Desesrt","-","%","%","%"],
        ["%","%", "%", "%", "%","-", "Desesrt", "Desesrt","Desesrt","Desesrt","-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Academic","Academic","Academic","Academic","Academic","Academic","Desesrt","Desesrt","Gardens","Gardens","-","Gardens","Gardens","Desesrt","-","Desesrt","Desesrt","Desesrt","-","%","%","%"],
        ["%","%", "%", "%", "%","-", "Desesrt", "Desesrt","Desesrt","Desesrt","Desesrt","-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","-","-","-","-","-","-","Desesrt","Desesrt","Gardens","Gardens","-","Gardens","Gardens","Desesrt","-","Desesrt","Desesrt","Desesrt","-","%","%","%"],
        ["%","%", "%", "%", "%","-", "Desesrt", "Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","-","Adminstration","Adminstration","Adminstration","-","-","Desesrt","Desesrt","Gardens","Gardens","-","Gardens","Gardens","Desesrt","-","Desesrt","Desesrt","Desesrt","-","%","%","%"],
        ["%","%", "%", "%", "%","-", "Desesrt", "Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Adminstration","Adminstration","Adminstration","Adminstration","Adminstration","-","Desesrt","Desesrt","Gardens","Gardens","-","Gardens","Gardens","Desesrt","-","Desesrt","Desesrt","Desesrt","-","%","%","%"],
        ["%","%", "%", "%", "%","%", "-", "Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Nano","Nano","-","Desesrt","Desesrt","Desesrt","Desesrt","Adminstration","Adminstration","Adminstration","Adminstration","Adminstration","Adminstration","Adminstration","-","Desesrt","Helmy","Helmy","-","Gardens","Gardens","Desesrt","-","Desesrt","Desesrt","Desesrt","-","%","%","%"],
        ["%","%", "%", "%", "%","%", "%", "-","Desesrt","Desesrt","Desesrt","Desesrt","Nano","Nano","Nano","-","Desesrt","Desesrt","Adminstration","Adminstration","Adminstration","Adminstration","Adminstration","Adminstration","Adminstration","Adminstration","Adminstration","-","Gardens","Helmy","Helmy","Gardens","Gardens","Desesrt","-","Desesrt","Desesrt","Desesrt","-","%","%","%"],
        ["%","%", "%", "%", "%","%", "%", "-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","Nano","Nano","Nano","-","Adminstration","Adminstration","Adminstration","Adminstration","Adminstration","Adminstration","Adminstration","Adminstration","Adminstration","Adminstration","Adminstration","-","Gardens","Helmy","Helmy","Gardens","Desesrt","-","%","%","%","%","%","%","%"],
        ["%","%", "%", "%", "%","%", "%", "%","-","Desesrt","Desesrt","Desesrt","Desesrt","OneStop","-","Nano","Nano","Nano","-","Adminstration","Adminstration","Adminstration","Adminstration","Adminstration","Adminstration","Adminstration","-","-","-","Helmy","Helmy","Helmy","Helmy","Desesrt","-","%","%","%","%","%","%","%"],
        ["%","%", "%", "%", "%","%", "%", "%","%","-","Desesrt","Desesrt","OneStop","OneStop","-","Nano","Nano","Nano","-","-","-","-","-","-","-","-","-","-","-","Gardens","Helmy","Helmy","Helmy","Helmy","-","%","%","%","%","%","%","%"],
        ["%","%", "%", "%", "%","%", "%", "%","%","%","-","OneStop","OneStop","-","Nano","Nano","Nano","Nano","-","-","-","-","-","-","-","-","-","-","-","Gardens","-","Helmy","Helmy","Desesrt","-","%","%","%","%","%","%","%"],
        ["%","%", "%", "%", "%","%", "%", "%","%","%","%","-","Desesrt","Desesrt","Desesrt","Desesrt","-","Desesrt","-","Desesrt","-","-","Desesrt","-","Desesrt","-","-","-","-","Gardens","-","Gardens","Gardens","Desesrt","-","%","%","%","%","%","%","%"],
        ["%","%", "%", "%", "%","%", "%", "%","%","%","%","%","-","Desesrt","-","Desesrt","-","Desesrt","-","-","-","Desesrt","-","Desesrt","-","Desesrt","-","-","-","Gardens","-","Gardens","Gardens","Desesrt","-","%","%","%","%","%","%","%"],
        ["%","%", "%", "%", "%","%", "%", "%","%","%","%","%","%","-","Desesrt","-","-","-","Desesrt","-","Desesrt","-","Desesrt","-","-","-","-","-","-","Gardens","-","Gardens","Gardens","Desesrt","-","%","%","%","%","%","%","%"],
        ["%","%", "%", "%", "%","%", "%", "%","%","%","%","%","%","%","-","Desesrt","-","-","-","Desesrt","-","Desesrt","-","-","Desesrt","-","-","-","-","Gardens","-","Gardens","Gardens","Desesrt","-","%","%","%","%","%","%","%"],
        ["%","%", "%", "%", "%","%", "%", "%","%","%","%","%","%","%","%","-","-","-","-","Desesrt","Desesrt","-","-","-","Desesrt","-","Desesrt","-","-","Gardens","-","Gardens","Gardens","Desesrt","-","%","%","%","%","%","%","%"],
        ["%","%", "%", "%", "%","%", "%", "%","%","%","%","%","%","%","%","%","-","-","-","Desesrt","Desesrt","-","-","-","Desesrt","-","-","-","-","Gardens","-","Gardens","Gardens","Desesrt","-","%","%","%","%","%","%","%"],
        ["%","%", "%", "%", "%","%", "%", "%","%","%","%","%","%","%","%","%","%","%","-","-","-","-","-","-","Desesrt","-","Desesrt","-","-","Desesrt","Desesrt","Desesrt","Desesrt","Desesrt","-","%","%","%","%","%","%","%"],
        ["%","%", "%", "%", "%","%", "%", "%","%","%","%","%","%","%","%","%","%","%","-","-","-","-","-","-","-","-","-","-","-","Desesrt","Desesrt","Desesrt","Desesrt","-","%","%","%","%","%","%","%","%"],
        ["%","%", "%", "%", "%","%", "%", "%","%","%","%","%","%","%","%","%","%","%","%","%","%","MainGate", "%", "%", "%","%", "%", "%","%","%","%","%","%","%","%","%","%","%","%","%","%","%"],
        ["%","%", "%", "%", "%","%", "%", "%","%","%","%","%","%","%","%","%","%","%","%","%","%","%", "%", "%", "%","%", "%", "%","%","%","%","%","%","%","%","%","%","%","%","%","%","%"]
        ]

def Reward(Idx,Item):
  #print('ZCMap[Idx[0]][Idx[1]]',ZCMap[Idx[0]][Idx[1]],'Idx' ,Idx)

  if ZCMap[Idx[0]][Idx[1]] == "%" : 
    return -5000;
  elif ZCMap[Idx[0]][Idx[1]] == "Desesrt" : return -2000
  elif ZCMap[Idx[0]][Idx[1]] == "Gardens" : return -1000
  elif ZCMap[Idx[0]][Idx[1]] == "-" and not Item: return -1
  elif ZCMap[Idx[0]][Idx[1]] == "-" and Item: return 5
  elif ZCMap[Idx[0]][Idx[1]] == "Dorms" : return -20
  elif ZCMap[Idx[0]][Idx[1]] == "Adminstration" : return -20
  elif ZCMap[Idx[0]][Idx[1]] == "Service" : return -20
  elif ZCMap[Idx[0]][Idx[1]] == "MainGate" : return -10
  elif ZCMap[Idx[0]][Idx[1]] == "Academic" and Item : return 30000
  elif ZCMap[Idx[0]][Idx[1]] == "Academic" and not Item : return -3000
  elif ZCMap[Idx[0]][Idx[1]] == "OneStop" and Item : return 30000
  elif ZCMap[Idx[0]][Idx[1]] == "OneStop" and not Item : return -3000
  elif ZCMap[Idx[0]][Idx[1]] == "Nano" and not Item : return 30000
  elif ZCMap[Idx[0]][Idx[1]] == "Nano" and Item : return -3000
  elif ZCMap[Idx[0]][Idx[1]] == "Helmy" and not Item : return 30000
  elif ZCMap[Idx[0]][Idx[1]] == "Helmy" and Item : return -3000
  #else: print('a3333333333333333333333333333333333333')
  else: 
    print('Idx',Idx)
    print('llllllllla3333333333333333333333333333333333333')

    return -1
class DynamicAgenet(Environment):
    '''Dynamic Agent world'''
    
    def __init__(self, Agent,item):
        self.Item = item
        self.no_item=8
        self.HB_items=4
        self.NB_items=4
        self.OSS_items=0
        self.AB_items=0
        self.HB=(25,28)
        self.AB=(12,21)
        self.OSS=(28,13)
        self.NB=(25,12)
        self.Main_Gate=(39, 21)
        self.discount = 0.8       
        self.bounds=(41,42) #bounds=(colum, row)
        self.state = (Agent,item)



    def actions(self):
        if self.state is None: return []

        elif (self.state[0] == self.NB) and (self.NB_items>0) and(self.state[1]==0):
            
            return['Pick']
 
        elif (self.state[0] == self.HB) and (self.HB_items>0) and(self.state[1]==0):

            return['Pick']

        elif (self.state[0] == self.OSS) and (self.OSS_items<4) and (self.state[1]==1):
            
            return['Drop']

        elif (self.state[0] == self.AB) and (self.AB_items<4) and (self.state[1]==1):
              
            return['Drop']

        else:
             return ['up','up-right','up-left', 'down','down-right','down-left', 'left', 'right']
    

    
    def apply(self, action):
        up = lambda position: (max(position[0] - 1, 0),position[1])
        down = lambda position: (  min(position[0] + 1, self.bounds[0] - 1),position[1])
        left = lambda position: ( position[0],max(position[1] - 1, 0))
        right = lambda position: ( position[0],min(position[1] + 1, self.bounds[1] - 1))
        up_right= lambda position: (  max(position[0] - 1, 0), min(position[1] + 1, self.bounds[1] - 1))
        up_left= lambda position: (max(position[0] - 1, 0),max(position[1] - 1, 0))
        down_right= lambda position: ( min(position[0] + 1, self.bounds[0] - 1),min(position[1] + 1, self.bounds[1] - 1))
        down_left= lambda position: (  min(position[0] + 1, self.bounds[0] - 1),max(position[1] - 1, 0))

        Agent, item = self.state
     #   print('zzz',action)
        if action == 'Drop':
           if(Agent==self.OSS):
             self.OSS_items=self.OSS_items+1
           elif(Agent==self.AB):
         #    print('still')
             self.AB_items=self.AB_items+1  
           item=0
           Agent = self.Main_Gate #Main Gate

           if (self.OSS_items==4 and self.AB_items==4):
             self.state = None           
           else:
             self.state = (Agent,item)

          # print("Drop")
           return Reward(Agent,item)

        elif action ==  'Pick':
          if(Agent==self.HB):
             self.HB_items=self.HB_items-1
          elif(Agent==self.NB):
             self.NB_items=self.NB_items-1            

          item=1
          self.state = (Agent,item) 
 #         print("Pick")
          return Reward(Agent,item)
        elif   action =='up': 
          self.state = (up(Agent),item) 
          return Reward(up(Agent),item)

        elif action ==   'down': 
          self.state = (down(Agent),item) 
          return Reward(down(Agent),item)

        elif action ==   'left': 
          self.state = (left(Agent),item) 
          return Reward(left(Agent),item) 

        elif action ==  'right': 
          self.state = (right(Agent),item) 
          return Reward(right(Agent),item) 

        elif action == 'up-right': 
          self.state=(up_right(Agent),item) 
          return Reward(up_right(Agent),item)

        elif action == 'up-left': 
          self.state=(up_left(Agent),item) 
          return Reward(up_left(Agent),item)

        elif action == 'down-right': 
          self.state=(down_right(Agent),item) 
          return Reward(down_right(Agent),item)

        elif action == 'down-left': 
          self.state=(down_left(Agent),item) 
          return Reward(down_left(Agent),item)

        else:

          return -1
        



_visualizers[DynamicAgenet] =  _ourAgent_visualizer
```
